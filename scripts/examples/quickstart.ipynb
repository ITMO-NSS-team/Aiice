```python
from pprint import pprint
import os

import torch.nn as nn
```

# Quickstart

```python
from aiice import AIICE
```

Let's take a simple persistence model that repeats the last observed frame

```python
class DummyModel(nn.Module):
    def __init__(self, forecast_len: int):
        super().__init__()
        self.forecast_len = forecast_len

    def forward(self, x):
        # X.shape = [B, pre_history_len, H, W]
        # Y.shape = [B, forecast_len, H, W]
        return x[:, -1:, :, :].repeat(1, self.forecast_len, 1, 1)

model = DummyModel(forecast_len=6)
```

AIICE loads the dataset, prepares sliding windows and evaluates the model.

```python
aiice = AIICE(
    pre_history_len=12,
    forecast_len=6,
    batch_size=4,
    start="2024-01-01",
    end="2024-04-01",
)
```

```python
results = aiice.bench(
    model,
    path="./data"  # optional: saves forecast GIFs
)
```

```python
pprint(results)
```
